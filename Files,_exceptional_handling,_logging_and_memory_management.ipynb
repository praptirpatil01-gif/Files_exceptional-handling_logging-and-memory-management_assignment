Q6.What is the difference between import and from ... import in Python?

Ans:-

* import :-
When we use import, the entire module is loaded into the program, and we must use the module name to access its functions or variables.
Example: After import math, we call math.sqrt(10).
It keeps the code organized and avoids naming conflicts but requires longer syntax.

 * from … import :-
When we use from module import name, only the specific function or variable is imported, and we can use it directly without the module name.
Example: After from math import sqrt, we simply use sqrt(10).
It makes the code shorter but may cause conflicts if different modules have the same function name.

Q9.What is the difference between multithreading and multiprocessing?

Ans:-

* Multithreading :-

Multithreading uses multiple threads within the same process, which means all threads share the same memory space. It is useful for tasks that involve waiting, such as input/output operations. Multithreading is lightweight but can be limited by the Global Interpreter Lock (GIL) in Python.

 * Multiprocessing :-

Multiprocessing uses multiple processes , each with its own separate memory space , allowing true parallel execution. It is suitable for tasks that require heavy CPU usage. Multiprocessing avoids the GIL and provides better performance for CPU-intensive programs but consumes more memory.

Q18.What is the difference between os.fork() and multiprocessing in Python?

Ans:-
 * os.fork() :-

os.fork() is a low-level system call available mainly on Unix/Linux systems. It creates a child process by duplicating the current process. Both parent and child continue executing the same code after the fork. It is fast but less flexible and not available on Windows.

 * multiprocessing :-

The multiprocessing module is a high-level Python library that works on all major platforms (Windows, Linux, Mac). It allows creating processes using simple functions and classes, provides features like process pools, queues, and pipes, and handles data sharing safely. It is more powerful, portable, and easier to use.

```python
import logging


logging.basicConfig(
    filename="error.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)


try:
    a = 10
    b = 0
    result = a / b
except ZeroDivisionError as e:
    logging.error("Division by zero occurred!")
    print("Error logged successfully!")
```

Q16.Demonstrate how to use memory profiling to check the memory usage of a small program.

```python
import tracemalloc

# Start memory tracking
tracemalloc.start()

# Small program
numbers = [i for i in range(5000)]
total = sum(numbers)
print("Sum =", total)

# Get memory usage
current, peak = tracemalloc.get_traced_memory()
print(f"Current memory usage: {current / 1024:.2f} KB")
print(f"Peak memory usage: {peak / 1024:.2f} KB")

# Stop tracking
tracemalloc.stop()
```

Q18. How would you implement a basic logging setup that logs to a file with rotation after 1MB.

```python
import logging
from logging.handlers import RotatingFileHandler

handler = RotatingFileHandler("rotate.log", maxBytes=1_000_000, backupCount=3)
logging.basicConfig(handlers=[handler], level=logging.INFO)

logging.info("Logging with rotation enabled.")
```
